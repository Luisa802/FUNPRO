ESTRUCTURA DE REPETICIÓN MIENTRAS
Es una esctructura de repetición boolenana empleada cuando no saben cuantas veces se debe repetir la acción.

Comprende:


1.   Una condición inical
2.   Un centinela que evalua sis se cumple la condición final
3.   Un modificador de condición inicial.


```python
ahorro=int(input("Con cuánto dinero cuento?"))
opc=input("Desea comprar algo?")
while ahorro>0 and opc=="s":
  precio=int(input("Cuánto cuesta eso?"))
  if precio<ahorro:
    print("Lo compro!!")
    ahorro=ahorro-precio #acumulador
  else:
    print("No me alcanza")

  opc=input("Desea comprar algo?")
```
