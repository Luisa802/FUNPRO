#LISTAS
Son tipos de datos que cumplen con tres características
1. Finita: Se puede contar cuántos elementos tiene la lista
2. Homogénea: Sólo almacena un dato en específico
3. Ordenada o index: Sabemos qué dato ocupa cada posición

```python
n=10
s=0
m=0
sueldo=[]
for e in range(n):
  sueldo.append(int(input(f"Ingrese el sueldo del trabajador {e+1}:")))
  s=sueldo[e]+s
prom=s/n
print(f"El sueldo promedio de la empresa es: {prom}")
for e in range(n):
  if sueldo[e]>prom: m=m+1
print(f"{m} trabajadores gana más del promedio")
```

Ejercicio: 5 notas y se elimina la peor nota

```python
n=[]
for nota in range(1,6):
  n.append(float(input(f"Ingrese las notas {nota}: ")))
if n[0]<n[1]<n[2]<n[3]<n[4]:
  print(f"se va a eleminar la nota {n[0]}")
  n.remove(n[0])
elif n[1]<n[2]<n[3]<n[4]:
  print(f"se va a eleminar la nota {n[1]}")
  n.remove(n[1])
elif n[2]<n[3]<n[4]:
  print(f"se va a eleminar la nota {n[2]}")
  n.remove(n[2])
elif n[3]<n[4]:
  print(f"se va a eleminar la nota {n[3]}")
  n.remove(n[3])
else:
  print(f"se va a eliminar la nota {n[4]}")
  n.remove(n[4])
```
